```python
euro_rate = {
    "BG": bg_euro,
    "CZ": cz_euro,
    "DK": dk_euro,
    "HU": hu_euro,
    "RO": ro_euro,
    "SE": sw_euro,
    "UK": uk_euro,
}
```

```python
def convert_to_euro(row, column_name):
    if row['geo_code'] in euro_rate:
        return row[column_name] / euro_rate[row['geo_code']][row['year']]
    else:
        return row[column_name]
```

```python
intang_data = pd.read_csv('./data/intangibles analytical.csv', index_col=0)
intang_data = intang_data[intang_data['geo_code'].isin(EU_Country_codes)].reset_index(drop=True)
intang_data = intang_data[intang_data['year'] >= 1999].reset_index(drop=True)
intang_data['I_Intang'] = intang_data.apply(lambda row: convert_to_euro(row, 'I_Intang'), axis=1)
intang_data['I_Tang'] = intang_data.apply(lambda row: convert_to_euro(row, 'I_Tang'), axis=1)
```

```python
yearly_total_intangible = intang_data[intang_data['nace_r2_code'] == 'TOT'].groupby(['year'])['I_Intang'].sum().reset_index()
yearly_total_intangible['I_Intang_scaled'] = yearly_total_intangible['I_Intang'] / yearly_total_intangible['I_Intang'][0] * 100
```

```python
yearly_total_tangible = intang_data[intang_data['nace_r2_code'] == 'TOT'].groupby(['year'])['I_Tang'].sum().reset_index()
yearly_total_tangible['I_Tang_scaled'] = yearly_total_tangible['I_Tang'] / yearly_total_tangible['I_Tang'][0] * 100
```

```python
selected_countries_data.groupby(["geo_code", "year", "nace_r2_name"]).agg(
    {
        "geo_code": "first",
        "year": "first",
        "nace_r2_name": "first",
        "I_Intang": "sum",
        "I_Tang": "sum",
    }
).reset_index(drop=True)
```

```python
sectors_2010 = selected_countries_data[selected_countries_data['year'] == 2010].groupby(["geo_code", "nace_r2_name"]).agg(
    {
        "geo_code": "first",
        "nace_r2_code": "first",
        "nace_r2_name": "first",
        "I_Intang": "sum",
        "I_Tang": "sum",
    }
).reset_index(drop=True)

sectors_2021 = selected_countries_data[selected_countries_data['year'] == 2021].groupby(["geo_code", "nace_r2_name"]).agg(
    {
        "geo_code": "first",
        "nace_r2_code": "first",
        "nace_r2_name": "first",
        "I_Intang": "sum",
        "I_Tang": "sum",
    }
).reset_index(drop=True)
res = sectors_2021.copy()
res['I_Intang_diff'] = sectors_2021['I_Intang'] - sectors_2010['I_Intang']
res['I_Intang_diff_ratio'] = (sectors_2021['I_Intang'] / sectors_2010['I_Intang']) - 1
res['I_Tang_diff'] = sectors_2021['I_Tang'] - sectors_2010['I_Tang']
res
```

```python
uk_by_sector_change = res[res['geo_code'] == 'UK'].sort_values(by='I_Intang_diff', ascending=False)
uk_by_sector_change = uk_by_sector_change[uk_by_sector_change['nace_r2_code'].str.len() == 1]
top10_uk_sector_change = uk_by_sector_change.sort_values(by='I_Intang_diff', ascending=False).head(10)['nace_r2_name']
uk_by_sector_change
```
